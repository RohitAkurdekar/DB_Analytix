```python
import os
os.environ['PYSPARK_SUBMIT_ARGS'] = '--packages org.apache.spark:spark-streaming-kafka-0-8_2.11:2.0.2 pyspark-shell'
```

```python
import findspark
findspark.init('/usr/local/spark')

from pyspark import SparkContext
from pyspark.streaming import StreamingContext
from pyspark.streaming.kafka import KafkaUtils
```

```python
sc = SparkContext("local[2]","Assessment")
ssc = StreamingContext(sc,5)
```

```python
kafkaStream = KafkaUtils.createStream(ssc,'localhost:2181','kafka-spark-streaming',{'input':1})
```

```python
lines = kafkaStream.map(lambda x:x[1])
```

```python
upperWords=lines.map(lambda word:word.upper())
```

```python
lowerWords=upperWords.map(lambda word:word.lower())

```

```python
counts = lowerWords.flatMap(lambda line: line.split(' ')).map(lambda word:(word,1)).reduceByKey(lambda a,b:a+b)
```

```python
upperWords.pprint()
lowerWords.pprint()
counts.pprint()
```
